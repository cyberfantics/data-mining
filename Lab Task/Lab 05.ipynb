</span>
</center>
</div>

```python
threshold = 0.5  # 50% threshold
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().mean()
df_cleaned = df.loc[:, missing_percentage < threshold]
```

```python
# Check the shape after removing columns
print(f"Original shape: {df.shape}")
print(f"Shape after removing high-missing columns: {df_cleaned.shape}")
```

```python
df = df_cleaned
df.isna().sum()
```

<div style="font-family: 'Arial', sans-serif; font-size: 20px; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 3px solid #555555; border-radius: 15px; box-shadow: 0 6px 15px rgba(0, 0, 0, 0.4); margin-bottom: 25px;">
<center>
  <span style="color:#4682B4; font-size:22px; font-weight: bold; padding: 5px;">Fill missing values for numerical columns with the mean</span>
</center>
</div>

```python
df.loc[:, 'comp.ratio'] = df['comp.ratio'].fillna(df['comp.ratio'].mean())
```

```python
df.dtypes
```

<div style="font-family: 'Arial', sans-serif; font-size: 20px; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 3px solid #555555; border-radius: 15px; box-shadow: 0 6px 15px rgba(0, 0, 0, 0.4); margin-bottom: 25px;">
<center>
  <span style="color:#4682B4; font-size:22px; font-weight: bold; padding: 5px;">Fill Low Missing Value with Mode</span>
</center>
</div>

```python
df.loc[:, 'transmission'] = df['transmission'].fillna(df['transmission'].mode()[0])
df.loc[:, 'ignition'] = df['ignition'].fillna(df['ignition'].mode()[0])
df.loc[:, 'wheelbase'] = df['wheelbase'].fillna(df['wheelbase'].mode()[0])
df.loc[:, 'seat.height'] = df['seat.height'].fillna(df['seat.height'].mean())
df.loc[:, 'clearance'] = df['clearance'].fillna(df['clearance'].mean())
df.loc[:, 'fuel.cap'] = df['fuel.cap'].fillna(df['fuel.cap'].mode()[0])
df.loc[:, 'total.weight'] = df['total.weight'].fillna(df['total.weight'].mode()[0])
```
