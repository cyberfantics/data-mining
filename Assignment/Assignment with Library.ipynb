```python
# Step 1: Data Cleaning (removing URLs, special characters, etc.)
def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|https\S+', '', text)
    # Remove hashtags and mentions
    text = re.sub(r'[@#]\S+', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove any special characters
    text = re.sub(r'[^a-z\s]', '', text)
    return text
```

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: center; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 3px solid #555555; border-radius: 15px; box-shadow: 0 6px 15px rgba(0, 0, 0, 0.4); margin-bottom: 25px;">
    <strong style="color: #ffcc00;"> Clean The Entire Dataset</strong>
</div>

```python
cleaned_text = [clean_text(doc) for doc in text]
```

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: center; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 3px solid #555555; border-radius: 15px; box-shadow: 0 6px 15px rgba(0, 0, 0, 0.4); margin-bottom: 25px;">
    <strong style="color: #ffcc00;"> Tokenize using NLTK</strong>
</div>

```python
tokenized_text = [word_tokenize(doc) for doc in cleaned_text]
```

```python
for text in tokenized_text:
    for sub in text:
        print(text)
```

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: center; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 3px solid #555555; border-radius: 15px; box-shadow: 0 6px 15px rgba(0, 0, 0, 0.4); margin-bottom: 25px;">
    <strong style="color: #ffcc00;"> Perform TF (Term Frequency) using TfidfVectorizer </strong>
</div>

```python
# We use TfidfVectorizer to calculate the TF for unigrams.
vectorizer = TfidfVectorizer(analyzer='word', ngram_range=(1, 1), stop_words='english')
```

```python
# Fit and transform the cleaned text to get the TF values
X = vectorizer.fit_transform(cleaned_text)
```

<div style="font-family: 'Arial', sans-serif; font-size: 20px; text-align: center; color: #ffffff; background-color: #2d2d2d; padding: 25px; border: 3px solid #555555; border-radius: 15px; box-shadow: 0 6px 15px rgba(0, 0, 0, 0.4); margin-bottom: 25px;">
    <strong style="color: #ffcc00;"> Convert the DataFrame for better visualization</strong>
</div>

```python
# Get feature names (terms)
terms = vectorizer.get_feature_names_out()
```

```python
# Convert the result to a pandas DataFrame for easy visualization
df = pd.DataFrame(X.toarray(), columns=terms)
```
